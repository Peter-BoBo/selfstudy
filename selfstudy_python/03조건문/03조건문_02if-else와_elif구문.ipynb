# if~else와 elif 구문

* if 조건문은 뒤에 else 구문을 붙여서 사용할 수 있다.

## *else 조건문의 활용
* if 조건문 뒤에 사용하며, if 조건문의 조건이 거짓일 때 실행되는 부분.

if 조건:<br>
□□□□조건이 참일 때 실행할 문장

else:<br>
□□□□조건이 거짓일 때 실행할 문장

```python
# if 조건문에 else 구문을 추가해서 짝수와 홀수 구분하기

number = input("정수 입력> ")
number = int(number)

if number %2 == 0:
    print("짝수입니다")
else:
    print("홀수입니다")
```

## *elif 구문

* 세 개 이상의 조건을 연결해서 사용하는 방법

if 조건A:<br>
□□□□조건A가 참일 때 실행할 문장

elif 조건B:<br>
□□□□조건B가 참일 때 실행할 문장

elif 조건C:<br>
□□□□조건C가 참일 때 실행할 문장

. . . 

else:<br>
□□□□모든 조건이 거짓일 때 문장

```python
# 날짜/시간과 관련된 기능을 가져옵니다.
import datetime

now = datetime.datetime.now()
month = now.month

if 3 <= month <= 5:
    print("현재는 봄입니다")
    
elif 6 <= month <= 8:
    print("현재는 여름입니다")

elif 9 <= month <= 11:
    print("현재는 가을입니다")
    
else:
    print("현재는 겨울입니다")
```

```python
# 날짜/시간과 관련된 기능을 가져옵니다.
import datetime

month = 12

if 3 <= month <= 5:
    print("현재는 봄입니다")
    
elif 6 <= month <= 8:
    print("현재는 여름입니다")

elif 9 <= month <= 11:
    print("현재는 가을입니다")
    
else:
    print("현재는 겨울입니다")
```

## if 조건문을 효율적으로 사용하기

```python
# 유머를 조건문으로 구현하기
score = float(input("학점 입력> "))

if score == 4.5:
    print("신")
elif 4.2 <= score < 4.5:
    print("교수님의 사랑")
elif 3.5 <= score < 4.2:
    print("현 체제의 수호자")
elif 2.8 <= score < 3.5:
    print("일반인")
elif 2.3 <= score < 2.8:
    print("일탈을 꿈꾸는 소시민")
elif 1.75 <= score < 2.3:
    print("오락문화의 선구자")
elif 1.0 <= score < 1.75:
    print("불가촉천민")
elif 0.5 <= score < 1.0:
    print("자벌레")
elif 0 < score < 0.5:
    print("플랑크톤")
elif 0 == score:
    print("시대를 앞서가는 혁명의 씨앗")
```

```python
# 유머를 조건문으로 구현하기(2)
score = float(input("학점 입력> "))

if score == 4.5:
    print("신")
elif 4.2 <= score:
    print("교수님의 사랑")
elif 3.5 <= score:
    print("현 체제의 수호자")
elif 2.8 <= score:
    print("일반인")
elif 2.3 <= score:
    print("일탈을 꿈꾸는 소시민")
elif 1.75 <= score:
    print("오락문화의 선구자")
elif 1.0 <= score:
    print("불가촉천민")
elif 0.5 <= score:
    print("자벌레")
elif 0 < score:
    print("플랑크톤")
else:
    print("시대를 앞서가는 혁명의 씨앗")
```
