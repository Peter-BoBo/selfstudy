# 범위 자료형과 while 반복문

* 어떤 대상을 기반으로 반복을 적용하는 것이 아닌,<br> **'특정 횟수만큼 반복한다', '특정 시간 동안 반복한다', '어떤 조건이 될 때까지 반복한다'와 같이 다양한 조건에 따라 반복하는 것이 범위 자료형**

## 범위
* 정수로 이루어진 범위를 만들 때는 range()함수를 사용한다.

#### 첫째, 매개변수에 숫자 한 개를 넣는 방법
0부터 A-1 까지의 정수로 범위를 만든다. <br>
range(A) -> A는 숫자

```python
a = range(5)
a
```

```python
list(range(5))
```

#### 둘째, 매개변수에 숫자 두 개를 넣는 방법
A부터 B-1 까지의 정수로 범위를 만든다. <br>
range(A, B) -> A와 B는 숫자

```python
list(range(0,5))
```

```python
list(range(5,10))
```

#### 셋째, 매개변수에 숫자 세 개를 넣는 방법
A부터 B-1 까지의 정수로 범위를 만드는데, 앞뒤의 숫자가 C만큼의 차이를 가진다.<br>
range(A, B, C) -> A, B, C는 숫자

```python
list(range(0, 10, 2))
```

```python
list(range(0, 10, 3))
```

```python
# 매개변수 내부에 수식을 사용하는 경우

a = range(0, 10 + 1)
list(a)
```

```python
n = 10
a = range(0, (n // 2)) #정수 나누기 연산자를 사용한다

list(a)
```

## *for 반복문: 범위와 함께 사용하기

for 숫자 변수 in 범위:<br>
&emsp;&nbsp;코드

```python
# for 반복문과 범위

for i in range(5):
    print(str(i) + "= 반복 변수")
print()

for i in range(5, 10):
    print(str(i) + "= 반복 변수")
print()

for i in range(0, 10, 3):
    print(str(i) + "= 반복 변수")
print()
```

## for 반복문: 리스트와 범위 조합하기

```python
# 리스트와 범위를 조합해서 사용하기
array = [273, 32, 103, 57, 52]

for i in range(len(array)):
    print(f'{i}번째 반복: {array[i]}')
```

## for 반복문: 반대로 반복하기
* 작은 숫자에서 큰 숫자 순서가 아닌, 큰 숫자에서 작은 숫자로 (역반복문)

```python
# 반대로 반복하기(1)

for i in range(4, -1, -1):
    print(f'현재 반복 변수: {i}')
```

```python
# 반대로 반복하기(2)

for i in reversed(range(5)):
    print(f"현재 반복 변수: {i}")
```

## 중첩 반복문으로 피라미드 만들기

```python
# 반복문으로 피라미드 만들기(1)

output = ""

for i in range(1, 10):
    for j in range(0, i):
        output += "*"
    output += "\n"

print(output)
```

```python
# 반복문으로 피라미드 만들기(2)

output = ""

for i in range(1, 15):
    for j in range(14, i, -1):
        output += ' '
    for k in range(0, 2 * i - 1):
        output += '*'
    output += "\n"
    
print(output)
```

```python
# 문자열 * 연산자로 피라미드 만들기

output = ""

for i in range(1,10):
    output += ("*" * i)
    output += "\n"
    
print(output)
```

## while 반복문

* <불 표현식>이 참인 동안 문장을 계속 반복한다.
* '무한 반복'이 구현 가능하다
* '조건'을 활용해서 반복을 사용해야 한다면 while 반복문을 사용한다.

whilke 불 표현식:<br>
&emsp;&nbsp;문장

```python
# 무한 반복

while True:
    print(".", end="")
```

## while 반복문: for 반복문처럼 사용하기

```python
# while 반복문을 for 반복문처럼 사용하기

i = 0
while i < 10:
    print(f'{i}번째 반복입니다.')
    i += 1
```

## while 반복문: 상태를 기반으로 반복하기

```python
# 해당하는 값 모두 제거하기

list_test = [1, 2, 1, 2]
value = 2

# list_test 내부에 value가 있다면 반복
while value in list_test:
    list_test.remove(value)
    
print(list_test)
```

## while 반복문: 시간을 기반으로 반복하기

* 시간을 기반으로 반복하려면 '유닉스 타임'이라는 개념을 알아야 한다.
* 유닉스 타임: 1970년 1월 1일 0시 0분 0초를 기준으로 몇 초가 지났는지를 정수로 나타낸 것

```python
# 5초 동안 반복하기

import time

time.time()

number = 0

target_tick = time.time() + 5
while time.time() < target_tick:
    number += 1
    
print(f'5초 동안 {number}번 반복했습니다.')
```

## while 반복문: break 키워드/continue 키워드

* break: 반복문을 벗어날 때 사용하는 키워드. (무한 반복을 벗어날 때 많이 사용)
* continue: 현재 반복을 생략하고, 다음 반복으로 넘어갈 때 사용하는 키워드

```python
# break 키워드

i = 0

while True:
    print(f'{i}번째 반복문입니다.')
    i = i + 1
    input_text = input(">종료하시겠습니까?(y/n): ")
    if input_text in ["y", "Y"]:
        print("반복을 종료합니다.")
        break
```

```python
# continue 키워드

numbers = [5, 15, 6, 20, 7, 25]

for number in numbers:
    if number < 10:
        continue
    print(number)
```
